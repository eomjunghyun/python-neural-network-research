```python
from pathlib import Path
import sys

def find_repo_root(start: Path | None = None) -> Path:
    current = Path.cwd().resolve() if start is None else Path(start).resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "src" / "config.py").exists() and (candidate / "README.md").exists():
            return candidate
    raise RuntimeError("Could not find the repository root containing src/config.py")

REPO_ROOT = find_repo_root()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

print(f"REPO_ROOT = {REPO_ROOT}")
```

```python
EXPERIMENT_NAME = "20260406_k_sweep_k01_initial"

# Do not change anything here unless you intentionally want to change
# a hyperparameter other than NUM_FREQS.
FIXED_CONFIG = dict(
    GLOBAL_SEED=42,
    DATA_SEED_BASE=1000,
    NOISE_SEED_BASE=50000,
    TRAIN_SEED_BASE=100,
    SEQ_LEN=1000,
    time_mode="continuous",
    DT=0.05,
    theta_min=0.05 * np.pi,
    theta_max=0.85 * np.pi,
    NUM_FREQS_MIN=1,
    NUM_FREQS_MAX=30,
    FREQ_MIN=1,
    FREQ_MAX=60,
    NYQUIST_MARGIN=0.98,
    RANDOM_AMPLITUDE=True,
    AMP_MIN=0.5,
    AMP_MAX=2.0,
    RANDOM_PHASE=False,
    PHASE_MIN=0.0,
    PHASE_MAX=2.0 * np.pi,
    USE_NOISE=False,
    NOISE_TYPE="white",
    SNR_DB=10.0,
    AR1_RHO=0.8,
    IMPULSE_PROB=0.01,
    IMPULSE_SCALE=8.0,
    TRAIN_TARGET="noisy",
    MODEL_ID="AN001_BN_RELU",
    LAG=32,
    HIDDEN_DIM=64,
    BOTTLENECK_MULTIPLIER=4,
    LR=0.01,
    EPOCHS=1000,
    NUM_EXPERIMENTS=5,
    SEEDS_PER_FREQ=10,
    TEST_RATIO=0.2,
    ACC_TOLERANCE=0.1,
    RANK_THRESHOLD=0.05,
    SCREE_TOPK=20,
    NORMALIZE_H_COLUMNS=False,
    VERBOSE=True,
    MAKE_PLOTS=False,
)

pd.DataFrame(list(FIXED_CONFIG.items()), columns=["hyperparameter", "value"])
```

```python
def _test_target_count(cfg: ExperimentConfig) -> int:
    n_targets_total = cfg.SEQ_LEN - cfg.LAG
    return max(1, int(np.floor(n_targets_total * cfg.TEST_RATIO)))


def _display_result_tables(cfg: ExperimentConfig, results: dict, k: int) -> None:
    summary_df = results["summary_df"]

    preferred_cols = [
        "set_idx",
        "min_delta_f",
        "min_delta_theta",
        "mse_mean",
        "test_mse_mean",
        "test_r2_mean",
        "rank_threshold_mean",
        "rank_gap_mean",
        "spectral_gap_2k_mean",
        "energy_ratio_2k_mean",
        "align_coverage_mean",
        "align_purity_mean",
        "alignment_score_2k_mean",
        "align_mean_cosine_mean",
        "mean_principal_angle_deg_mean",
        "fourier_numerical_dim_mean",
    ]
    display_cols = [col for col in preferred_cols if col in summary_df.columns]
    print(f"=== k={k} summary ===")
    display(summary_df[display_cols] if display_cols else summary_df)

    basis_rows = []
    n_test_targets = _test_target_count(cfg)
    for row in results["set_results"]:
        diag = calculate_sampled_basis_numerical_dim(
            time_mode=cfg.time_mode,
            freqs=row.get("freqs"),
            thetas=row.get("thetas"),
            dt=cfg.DT,
            lag=cfg.LAG,
            seq_len=n_test_targets,
        )
        basis_rows.append({
            "set_idx": row["set_idx"],
            "fourier_theoretical_dim": diag["fourier_theoretical_dim"],
            "fourier_numerical_dim": diag["fourier_numerical_dim"],
            "fourier_min_singular_value": diag["fourier_min_singular_value"],
            "fourier_condition_number": diag["fourier_condition_number"],
        })

    basis_df = pd.DataFrame(basis_rows)
    print(f"=== k={k} basis diagnostics (theory dim stays 2k; numerical rank is diagnostic only) ===")
    display(basis_df)


def run_k_experiment(k: int):
    cfg = ExperimentConfig(NUM_FREQS=k, **FIXED_CONFIG)
    results = run_experiment(cfg)
    _display_result_tables(cfg, results, k)
    return results
```
